```python

# importing OpenCV
import cv2
import matplotlib.pyplot as plt
```

```python
# Reading the image

image_add = '/home/vishnu/practice_codes/video_analysis_practice/video-analytics-e2e/OpenCV/images/640px-Akatsuki_clouds_shiny_tiles_texture.png'
image = cv2.imread(image_add, cv2.IMREAD_COLOR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# getting the height and width of image
height, width, channel = image.shape
print(height, width, channel)
```

```python
# resizing the image
resize = cv2.resize(image, (100, 100))
plt.imshow(resize)
resize.shape
```

```python
# drawing a rectangle in an Image
rect_image = image.copy()
top_left_corner = (200, 50)
bottom_right_corner = (550, 350)
color = (255, 0, 0)
line_width = 2
rectangle = cv2.rectangle(rect_image, top_left_corner, bottom_right_corner, color, line_width)
plt.imshow(rectangle)
```

```python
# adding a text on an image
text = image.copy()
text_to_display = 'AKATSUKI'
bottom_left_corner = (100, 400)
font = cv2.FONT_HERSHEY_PLAIN
font_size = 6
color = (0, 255, 0)
line_width = 4
text = cv2.putText(text, text_to_display, bottom_left_corner, font, font_size, color, line_width)
plt.imshow(text)
```

```python
# Edge Detection
edges = cv2.Canny(image, 100, 500)
plt.imshow(edges)
```
